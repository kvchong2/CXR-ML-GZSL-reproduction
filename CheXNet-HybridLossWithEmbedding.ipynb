# Embedding


```python

# Install transformers if not already installed
# !pip install transformers torch

import numpy as np
import torch
from transformers import AutoTokenizer, AutoModel

# Disease names (must match the exact order in your BioBERT embeddings)
disease_names = [
    'Atelectasis', 
    'Cardiomegaly', 
    'Effusion', 
    'Infiltration',
    'Mass', 
    'Nodule', 
    'Pneumonia', 
    'Pneumothorax',
    'Consolidation', 
    'Edema', 
    'Emphysema', 
    'Fibrosis',
    'Pleural_Thickening', 
    'Hernia'
]

print("Loading PubMedBERT model...")
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

# Load PubMedBERT
tokenizer = AutoTokenizer.from_pretrained("microsoft/BiomedNLP-PubMedBERT-base-uncased-abstract-fulltext")
model = AutoModel.from_pretrained("microsoft/BiomedNLP-PubMedBERT-base-uncased-abstract-fulltext").to(device)
model.eval()

print("Generating embeddings for 14 disease classes...")

# Generate embeddings
embeddings = []

with torch.no_grad():
    for i, disease in enumerate(disease_names):
        print(f"  [{i+1}/14] Processing: {disease}")
        # Tokenize disease name
        inputs = tokenizer(disease, return_tensors="pt", padding=True, truncation=True)
        inputs = {k: v.to(device) for k, v in inputs.items()}

        # Get embeddings from [CLS] token
        outputs = model(**inputs)
        embedding = outputs.last_hidden_state[:, 0, :].cpu().numpy()
        embeddings.append(embedding)

# Stack into [14, 768] array
pubmedbert_embeddings = np.vstack(embeddings)

print(f"PubMedBERT embeddings shape: {pubmedbert_embeddings.shape}")
print(f"  Expected: (14, 768)")

# Save in same format as BioBERT
np.save('embeddings/pubmedbert_embeddings.npy', pubmedbert_embeddings)
print(f"Saved to: embeddings/pubmedbert_embeddings.npy")

# Verify the file can be loaded
test_load = np.load('embeddings/pubmedbert_embeddings.npy')
print(f"✓ Verification: Successfully loaded embeddings with shape {test_load.shape}")

print("SUCCESS! PubMedBERT embeddings are ready to use.")

```

## Dataset Class (from dataset.py)

```python
class NIHChestXray(torch.utils.data.Dataset):
    """
    PyTorch Dataset class for NIH Chest X-ray dataset.
    Handles loading images and labels, with support for zero-shot learning splits.
    """
    def __init__ (self, args, pathDatasetFile, transform, classes_to_load='seen', exclude_all=True):
        """
        Args:
            args: Configuration object with dataset paths and settings
            pathDatasetFile: Path to text file containing list of image filenames for this split
            transform: PyTorch transforms to apply to images
            classes_to_load: 'seen', 'unseen', or 'all' - which classes to include
            exclude_all: If True, exclude images that have both seen and unseen classes
        """
        # Initialize empty lists for image paths and labels
        self.listImagePaths = []
        self.listImageLabels = []
        self.transform = transform  # Image transformations (resize, crop, normalize, etc.)
        self.num_classes = args.num_classes  # Total number of disease classes (14)

        # Store paths and arguments
        self._data_path = args.data_root  # Root directory of the dataset
        self.args = args
        self.split_path = pathDatasetFile  # Path to split file (train.txt, val.txt, test.txt)

        # Define all disease classes in the dataset
        self.CLASSES = [ 'Atelectasis', 'Cardiomegaly', 'Effusion', 'Infiltration', 'Mass', 'Nodule', 'Pneumonia',
                'Pneumothorax', 'Consolidation', 'Edema', 'Emphysema', 'Fibrosis', 'Pleural_Thickening', 'Hernia']

        # Zero-shot learning split: unseen classes (not seen during training)
        self.unseen_classes = ['Edema', 'Pneumonia', 'Emphysema', 'Fibrosis']

        # Zero-shot learning split: seen classes (available during training)
        self.seen_classes = [ 'Atelectasis', 'Effusion', 'Infiltration', 'Mass', 'Nodule',
                'Pneumothorax', 'Consolidation', 'Cardiomegaly', 'Pleural_Thickening', 'Hernia']

        # Create mapping from class name to class ID (0-13, excluding 'No Finding')
        self._class_ids = {v: i for i, v in enumerate(self.CLASSES) if v != 'No Finding'}

        # Convert class names to class IDs for seen and unseen classes
        self.seen_class_ids = [self._class_ids[label] for label in self.seen_classes]
        self.unseen_class_ids = [self._class_ids[label] for label in self.unseen_classes]

        # Store filtering options
        self.classes_to_load = classes_to_load  # Which classes to include in this dataset
        self.exclude_all = exclude_all  # Whether to exclude mixed-label images

        # Build the dataset index (loads all image paths and labels)
        self._construct_index()

    def _construct_index(self):
        """
        Build the dataset index by:
        1. Finding all image files in the dataset
        2. Reading the split file to get which images belong to this split
        3. Reading the CSV file to get labels for each image
        4. Filtering images based on class selection criteria
        """
        max_labels = 0  # Track maximum number of labels per image

        # Find all PNG images in the dataset directory structure
        paths = glob.glob(f'{self._data_path}/images/*.png')
        # Create mapping from filename to full path for quick lookup
        self.names_to_path = {path.split('/')[-1]: path for path in paths}

        # Read the data entry CSV file that contains image labels
        data_entry_file = 'Data_Entry_2017_v2020.csv'
        print(f'data partition path: {self.split_path}')

        # Read the split file (train.txt, val.txt, or test.txt) to get image filenames
        with open(self.split_path, 'r') as f:
            file_names = f.readlines()

        # Extract just the filename from each line (handles paths like "images/00000001_000.png")
        split_file_names = np.array([file_name.strip().split(' ')[0].split('/')[-1] for file_name in file_names])

        # Read the main CSV file containing all image labels
        df = pd.read_csv(f'{self._data_path}/{data_entry_file}')
        image_index = df.iloc[:, 0].values  # First column: image filenames

        # Find intersection: which images from the split file are in the CSV
        _, split_index, _ = np.intersect1d(image_index, split_file_names, return_indices=True)

        # Get labels for images in this split (second column of CSV contains labels)
        labels = df.iloc[:, 1].values
        labels = np.array(labels)[split_index]  # Filter labels to only this split

        # Labels are pipe-separated (e.g., "Atelectasis|Effusion"), split them into lists
        labels = [label.split('|') for label in labels]

        # Filter image index to only this split
        image_index = image_index[split_index]

        # Construct the image database: list of dictionaries with image path and labels
        self._imdb = []
        self.class_ids_loaded = []  # Track which class IDs appear in this dataset

        for index in range(len(split_index)):
            # Skip images with only "No Finding" label (no diseases)
            if len(labels[index]) == 1 and labels[index][0] == 'No Finding':
                continue

            # Apply class filtering (seen/unseen/all)
            if self._should_load_image(labels[index]) is False:
                continue

            # Convert class names to class IDs
            class_ids = [self._class_ids[label] for label in labels[index]]
            self.class_ids_loaded += class_ids  # Track all class IDs we've seen

            # Add image to database
            try:
              self._imdb.append({
                  'im_path': self.names_to_path[image_index[index]],  # Full path to image
                  'labels': class_ids,  # List of class IDs for this image
              })
              max_labels = max(max_labels, len(class_ids))  # Update max labels per image
            except:
              # Ignore in case the names are missing (didn't unpack all images) for testing...
              pass

        # Get unique class IDs that appear in this dataset
        self.class_ids_loaded = np.unique(np.array(self.class_ids_loaded))

        # Print dataset statistics
        print(f'Number of images: {len(self._imdb)}')
        print(f'Number of max labels per image: {max_labels}')
        print(f'Number of classes: {len(self.class_ids_loaded)}')


    def _should_load_image(self, labels):
        """
        Determine if an image should be included based on its labels and filtering criteria.

        Args:
            labels: List of class names for an image

        Returns:
            True if image should be loaded, False otherwise
        """
        # Determine which classes we're interested in based on classes_to_load
        selected_class_labels = self.CLASSES  # Default: all classes
        if self.classes_to_load == 'seen':
            selected_class_labels = self.seen_classes  # Only seen classes
        elif self.classes_to_load == 'unseen':
            selected_class_labels = self.unseen_classes  # Only unseen classes
        elif self.classes_to_load == 'all':
            return True  # Load all images regardless of labels

        # Count how many of this image's labels are in the selected set
        count = 0
        for label in labels:
            if label in selected_class_labels:
                count += 1

        if count == len(labels):
            # All labels are from the selected subset - include this image
            return True
        elif count == 0:
            # None of the labels are in the selected subset - exclude this image
            return False
        else:
            # Some labels are in the selected subset (mixed case)
            if self.exclude_all is True:
                return False  # Exclude images with mixed seen/unseen labels
            else:
                return True  # Include images with mixed labels


    def __getitem__(self, index):
        """
        Get a single image and its labels.

        Args:
            index: Index of the image to retrieve

        Returns:
            img: Transformed image tensor
            imageLabel: One-hot encoded label vector (14 classes)
        """
        # Get the image path from the database
        imagePath = self._imdb[index]['im_path']

        # Load and convert image to RGB (ensures 3 channels)
        imageData = Image.open(imagePath).convert('RGB')

        # Get class IDs for this image and convert to tensor
        labels = torch.tensor(self._imdb[index]['labels'])

        # Convert class IDs to one-hot encoding
        # First add a batch dimension, then create one-hot vector
        labels = labels.unsqueeze(0)  # Shape: [1, num_labels]
        # Create zero vector and scatter 1s at label positions
        imageLabel = torch.zeros(labels.size(0), self.num_classes).scatter_(1, labels, 1.).squeeze()
        # Result: [14] tensor with 1s at positions corresponding to diseases present

        # Apply transforms (resize, crop, normalize, etc.)
        img = self.transform(imageData)

        return img, imageLabel

    def __len__(self):
        """Return the number of images in the dataset."""
        return len(self._imdb)
```

## Loss Functions (from loss.py)

```python
class KLDivLoss(nn.Module):
    """
    Kullback-Leibler Divergence Loss.
    Measures the difference between two probability distributions.
    Used for knowledge distillation or distribution matching.
    """
    def __init__(self, temperature=0.2):
        """
        Args:
            temperature: Temperature parameter for softmax (lower = sharper distribution)
        """
        super(KLDivLoss, self).__init__()
        self.temperature = temperature

    def forward(self, emb1, emb2):
        """
        Compute KL divergence between two embeddings.

        Args:
            emb1: Reference embedding (detached, used as target distribution)
            emb2: Predicted embedding (learned, used as predicted distribution)

        Returns:
            KL divergence loss (scalar)
        """
        # Convert embeddings to probability distributions using temperature-scaled softmax
        emb1 = softmax(emb1/self.temperature, dim=1).detach()  # Target distribution (detached from graph)
        emb2 = log_softmax(emb2/self.temperature, dim=1)  # Predicted distribution (in log space)

        # Compute KL divergence: KL(P||Q) = sum(P * log(P/Q))
        loss_kldiv = kl_div(emb2, emb1, reduction='none')  # Element-wise KL divergence
        loss_kldiv = torch.sum(loss_kldiv, dim=1)  # Sum over classes for each sample
        loss_kldiv = torch.mean(loss_kldiv)  # Average over batch
        return loss_kldiv

class RankingLoss(nn.Module):
    """
    Ranking Loss for Zero-Shot Learning.
    Ensures that correct class similarities are higher than incorrect ones by a margin.
    This is the main loss function for training the ZSL model.
    """
    def __init__(self, neg_penalty=0.03):
        """
        Args:
            neg_penalty: Margin penalty - how much higher correct labels should rank than incorrect ones
        """
        super(RankingLoss, self).__init__()
        self.neg_penalty = neg_penalty

    def forward(self, ranks, labels, class_ids_loaded, device):
        """
        Compute ranking loss.
        For each image, ensures that similarity scores for present diseases are higher
        than scores for absent diseases by at least neg_penalty.

        Args:
            ranks: [batch_size, num_classes] similarity scores between images and class embeddings
            labels: [batch_size, num_classes] binary labels (1 = present, 0 = absent)
            class_ids_loaded: List of class IDs that are available in this dataset
            device: Device to place tensors on

        Returns:
            Ranking loss (scalar)
        """
        # Filter to only classes that are loaded in this dataset
        labels = labels[:, class_ids_loaded]  # [batch_size, num_loaded_classes]
        ranks_loaded = ranks[:, class_ids_loaded]  # [batch_size, num_loaded_classes]

        # Create negative labels (1 - labels): 1 where disease is absent, 0 where present
        neg_labels = 1 + (labels * -1)  # Invert labels

        loss_rank = torch.zeros(1).to(device)

        # Process each image in the batch
        for i in range(len(labels)):
            # Get similarity scores for present diseases (correct)
            correct = ranks_loaded[i, labels[i]==1]  # [num_correct]
            # Get similarity scores for absent diseases (wrong)
            wrong = ranks_loaded[i, neg_labels[i]==1]  # [num_wrong]

            # Create pairwise comparisons: for each correct class, compare with all wrong classes
            # Expand dimensions to create a matrix: [num_correct, num_wrong]
            correct = correct.reshape((-1, 1)).repeat((1, len(wrong)))  # [num_correct, num_wrong]
            wrong = wrong.repeat(len(correct)).reshape(len(correct), -1)  # [num_correct, num_wrong]

            # Compute penalty: (neg_penalty + wrong_score) - correct_score
            # We want correct_score > wrong_score + neg_penalty
            # So penalty = wrong_score + neg_penalty - correct_score should be <= 0
            image_level_penalty = ((self.neg_penalty + wrong) - correct)
            image_level_penalty[image_level_penalty < 0] = 0  # Only penalize when margin is violated

            # Sum all violations for this image
            loss_rank += image_level_penalty.sum()

        # Average over batch
        loss_rank /= len(labels)
        return loss_rank

class CosineLoss(nn.Module):
    """
    Cosine Similarity Loss.
    Measures the cosine similarity between two embeddings.
    Used to align visual and textual embeddings in the same space.
    """
    def forward(self, t_emb, v_emb):
        """
        Compute cosine similarity loss between text and visual embeddings.

        Args:
            t_emb: Text embeddings [batch_size, embedding_dim]
            v_emb: Visual embeddings [batch_size, embedding_dim]

        Returns:
            Cosine loss (1 - mean cosine similarity), ranges from 0 to 2
        """
        # Normalize embeddings to unit vectors
        a_norm = v_emb / v_emb.norm(dim=1)[:, None]  # [batch_size, embedding_dim]
        b_norm = t_emb / t_emb.norm(dim=1)[:, None]  # [batch_size, embedding_dim]

        # Compute cosine similarity matrix: [batch_size, batch_size]
        # Diagonal elements are similarities between corresponding pairs
        similarity_matrix = torch.mm(a_norm, b_norm.t())  # [batch_size, batch_size]
        diagonal_similarities = torch.diagonal(similarity_matrix, 0)  # [batch_size] - matching pairs

        # Loss is 1 minus mean cosine similarity (we want high similarity = low loss)
        loss = 1 - torch.mean(diagonal_similarities)
        return loss
```

# New loss function

```python
class HybridSemanticLoss(nn.Module):
    def __init__(self, alpha=0.7, temperature=0.1):
        super(HybridSemanticLoss, self).__init__()
        self.alpha = alpha
        self.temperature = temperature
        self.cosine_loss = CosineLoss()
        
    def forward(self, visual_feats, mapped_text_feats, original_text_feats):
        """
        Args:
            visual_feats: Output from self.fc_v [batch_size, 128]
            mapped_text_feats: Output from self.fc_t [num_classes, 128]
            original_text_feats: Your PubMedBERT embeddings [num_classes, 768]
        """
        # 1. Standard cosine loss between mapped features
        alignment_loss = self.cosine_loss(visual_feats, mapped_text_feats)
        
        # 2. Semantic preservation loss
        # Compare relationships in original vs mapped space
        orig_sim = F.cosine_similarity(
            original_text_feats.unsqueeze(1), 
            original_text_feats.unsqueeze(0), 
            dim=2
        )
        
        mapped_sim = F.cosine_similarity(
            mapped_text_feats.unsqueeze(1), 
            mapped_text_feats.unsqueeze(0), 
            dim=2
        )
        
        # Scale similarities
        orig_sim = torch.softmax(orig_sim / self.temperature, dim=1)
        mapped_sim = torch.softmax(mapped_sim / self.temperature, dim=1)
        
        # KL divergence between similarity distributions
        preservation_loss = F.kl_div(
            torch.log(mapped_sim + 1e-8),
            orig_sim,
            reduction='batchmean'
        )
        
        return self.alpha * alignment_loss + (1 - self.alpha) * preservation_loss
```

## ZSL Model (from zsl_models.py)

```python
class ZSLNetHybrid(nn.Module):
    """
    Zero-Shot Learning Network for Chest X-ray Disease Classification.

    This model combines:
    1. A vision backbone (e.g., DenseNet) to extract visual features from images
    2. Text embeddings (BioBERT) for disease class descriptions
    3. Mapping networks to project both modalities into a shared embedding space
    4. Similarity-based classification using cosine similarity between visual and text embeddings
    """
    def __init__(self, args, textual_embeddings=None, device='cpu'):
        """
        Initialize the ZSL network.

        Args:
            args: Configuration object with model hyperparameters
            textual_embeddings: Pre-computed text embeddings for disease classes [num_classes, embedding_dim]
            device: Device to run the model on ('cpu' or 'cuda')
        """
        super(ZSLNetHybrid, self).__init__()
        self.args = args
        self.device = device

        # Load pre-trained vision backbone (e.g., DenseNet121 with ImageNet weights)
        self.vision_backbone = getattr(torchvision.models, self.args.vision_backbone)(pretrained=self.args.pretrained)

        # Remove the classification layer from the vision backbone
        # We want to extract features, not classify into ImageNet classes
        classifiers = ['classifier', 'fc']  # Different models use different names
        for classifier in classifiers:
            cls_layer = getattr(self.vision_backbone, classifier, None)
            if cls_layer is None:
                continue  # Try next classifier name
            # Get the input dimension of the classifier (this is our visual feature dimension)
            d_visual = cls_layer.in_features
            # Replace classifier with identity function (pass-through)
            setattr(self.vision_backbone, classifier, nn.Identity())
            break  # Found and replaced, exit loop

        # Optional: Load a pre-trained encoder (currently disabled)
        pretrained_encoder = False
        if pretrained_encoder:
            self.vision_backbone.classifier = nn.Identity()
            path = 'checkpoints/bce_only_imagenet/last_epoch_checkpoint.pth.tar'
            self.classifier = nn.Sequential(nn.Linear(d_visual, self.args.num_classes), nn.Sigmoid())
            checkpoint = torch.load(path)
            self.load_state_dict(checkpoint['state_dict'])
            # Freeze vision backbone parameters
            for p in self.vision_backbone.parameters():
                p.requires_grad = False

        # Two training modes: BCE-only (standard classification) or ZSL (ranking-based)
        if self.args.bce_only:
            # Standard binary cross-entropy classification
            self.bce_loss = torch.nn.BCELoss(size_average=True)
            self.classifier = nn.Sequential(
                nn.Linear(d_visual, self.args.num_classes),
                nn.Sigmoid()  # Multi-label classification
            )
        else:
            # Zero-shot learning mode: use ranking loss and embedding alignment
            self.emb_loss = CosineLoss()  # Align visual and text embeddings
            self.ranking_loss = RankingLoss(neg_penalty=self.args.neg_penalty)  # Main ZSL loss
            # Add hybrid loss
            self.hybrid_loss = HybridSemanticLoss()
            # Store and convert text embeddings to tensor
            self.textual_embeddings = textual_embeddings
            d_textual = self.textual_embeddings.shape[-1]  # Dimension of text embeddings
            self.textual_embeddings = torch.from_numpy(self.textual_embeddings).to(self.device)
            self.original_embeddings = self.textual_embeddings.clone()  # Keep original embeddings
            
            # Visual feature mapping network: projects visual features to shared space
            # Architecture: d_visual -> 512 -> 256 -> 128
            self.fc_v = nn.Sequential(
                nn.Linear(d_visual, 512),
                nn.ReLU(),
                nn.Linear(512, 256),
                nn.ReLU(),
                nn.Linear(256, 128),  # Final visual embedding dimension
            )

            # Text feature mapping network: projects text embeddings to shared space
            # Architecture: d_textual -> 512 -> 256 -> 128
            self.fc_t = nn.Sequential(
                nn.Linear(d_textual, 512),
                nn.ReLU(),
                nn.Linear(512, 256),
                nn.ReLU(),
                nn.Linear(256, 128)  # Final text embedding dimension (matches visual)
            )


    def forward(self, x, labels=None, epoch=0, n_crops=0, bs=16):
        """
        Forward pass through the network.

        Args:
            x: Input images [batch_size, channels, height, width] or [batch_size*n_crops, ...] for test-time augmentation
            labels: Ground truth labels [batch_size, num_classes] (optional, for training)
            epoch: Current training epoch (used to enable/disable certain losses)
            n_crops: Number of crops per image (for test-time augmentation, 0 during training)
            bs: Batch size (needed when n_crops > 0 to reshape outputs)

        Returns:
            If bce_only: (predictions, loss, loss_string)
            Otherwise: (similarity_scores, total_loss)
        """
        if self.args.bce_only:
            return self.forward_bce_only(x, labels=labels, n_crops=n_crops, bs=bs)
        else:
            return self.forward_ranking(x, labels=labels, epoch=epoch, n_crops=n_crops, bs=bs)

    def forward_bce_only(self, x, labels=None, n_crops=0, bs=16):
        """
        Forward pass for BCE-only mode (standard multi-label classification).
        """
        lossvalue_bce = torch.zeros(1).to(self.device)

        # Extract visual features using the backbone
        visual_feats = self.vision_backbone(x)  # [batch_size, d_visual]
        # Classify directly to disease classes
        preds = self.classifier(visual_feats)  # [batch_size, num_classes]

        # Compute binary cross-entropy loss if labels provided
        if labels is not None:
            lossvalue_bce = self.bce_loss(preds, labels)

        return preds, lossvalue_bce, f'bce:\t {lossvalue_bce.item():0.4f}'


    def forward_ranking(self, x, labels=None, epoch=0, n_crops=0, bs=16):
        """
        Forward pass for ZSL ranking-based mode.
        Projects visual and text features to shared space and computes similarity scores.
        """
        # Initialize loss components
        loss_rank = torch.zeros(1).to(self.device)  # Main ranking loss
        loss_hybrid = torch.zeros(1).to(self.device)
        loss_allignment_cos = torch.zeros(1).to(self.device)  # Visual-text alignment loss
        loss_mapping_consistency = torch.zeros(1).to(self.device)  # Consistency regularizer

        # Extract visual features and project to shared embedding space
        visual_feats = self.vision_backbone(x)  # [batch_size, d_visual]
        visual_feats = self.fc_v(visual_feats)  # [batch_size, 128]

        # Project text embeddings to shared space
        text_feats = self.fc_t(self.textual_embeddings)  # [num_classes, 128]

        # Mapping consistency loss: preserve semantic relationships in text embeddings
        # Ensures that the mapped text embeddings maintain the same similarity structure as original
        if not self.args.wo_con and epoch >= 0:
            # Compute similarity matrix of mapped text embeddings
            text_mapped_sim = self.sim_score(text_feats, text_feats.detach())  # [num_classes, num_classes]
            # Compute similarity matrix of original text embeddings
            text_orig_sim = self.sim_score(self.textual_embeddings, self.textual_embeddings)  # [num_classes, num_classes]
            # Loss: difference between similarity matrices (should be small)
            loss_mapping_consistency = torch.abs(text_orig_sim - text_mapped_sim).mean()

        # Visual-text alignment loss: align visual features with corresponding text features
        # Currently disabled (multiplied by 0.0) but code is kept for potential use
        if labels is not None:
            mapped_visual, mapped_text = self.map_visual_text(visual_feats, labels, text_feats)
            if mapped_visual is not None and not self.args.wo_map and epoch >= 0:
                # Compute cosine loss between visual features and average text features of present classes
                loss_allignment_cos = self.emb_loss(mapped_text, mapped_visual)

        # Compute similarity scores between visual features and all class text embeddings
        ranks = self.sim_score(visual_feats, text_feats)  # [batch_size, num_classes]

        # Handle test-time augmentation: average predictions across multiple crops
        if n_crops > 0:
            # Reshape: [batch_size*n_crops, num_classes] -> [batch_size, n_crops, num_classes]
            ranks = ranks.view(bs, n_crops, -1).mean(1)  # Average over crops: [batch_size, num_classes]

        # Compute ranking loss if labels provided (training mode)
        if labels is not None:
            loss_rank = self.ranking_loss(ranks, labels, self.class_ids_loaded, self.device)
            if not self.args.wo_hybrid and epoch >= 0:  # Add similar control flag
                loss_hybrid = self.hybrid_loss(
                    visual_feats,
                    text_feats,
                    self.original_embeddings
                )

        # Weight and combine loss components
        loss_allignment_cos = (self.args.beta_map * loss_allignment_cos)  # Currently 0.0 weight
        loss_rank = (self.args.beta_rank * loss_rank)  # Main loss
        loss_mapping_consistency = (self.args.beta_con * loss_mapping_consistency)  # Regularizer
        loss_hybrid = (self.args.beta_hybrid * loss_hybrid)  # New weight for hybrid loss

        # Total loss (alignment loss disabled with 0.0 multiplier)
        losses = loss_rank + loss_mapping_consistency + 0.0 * loss_allignment_cos + loss_hybrid

        return ranks, losses

    def sim_score(self, a, b):
        """
        Compute cosine similarity matrix between two sets of embeddings.

        Args:
            a: First set of embeddings [N, embedding_dim]
            b: Second set of embeddings [M, embedding_dim]

        Returns:
            Similarity matrix [N, M] where element (i,j) is cosine similarity between a[i] and b[j]
        """
        # Normalize to unit vectors
        a_norm = a / a.norm(dim=1)[:, None]  # [N, embedding_dim]
        b_norm = b / (1e-6 + b.norm(dim=1))[:, None]  # [M, embedding_dim] (small epsilon for numerical stability)

        # Compute cosine similarity: dot product of normalized vectors
        score = torch.mm(a_norm, b_norm.t())  # [N, M]
        return score

    def map_visual_text(self, visual_feats, labels, labels_embd):
        """
        Map visual features to text embeddings for alignment loss.
        For each image, compute the average text embedding of its present classes.

        Args:
            visual_feats: Visual features [batch_size, embedding_dim]
            labels: Binary labels [batch_size, num_classes]
            labels_embd: Text embeddings for all classes [num_classes, embedding_dim]

        Returns:
            mapped_visual: Visual features (detached)
            mapped_labels_embd: Average text embeddings for present classes in each image [batch_size, embedding_dim]
        """
        mapped_labels_embd = []

        # Process each image in the batch
        for i in range(0, labels.shape[0]):
            # Get text embeddings for classes present in this image
            # labels[i]==1 gives boolean mask, use it to index into labels_embd
            class_embd = labels_embd[labels[i]==1].mean(dim=0)[None, :]  # [1, embedding_dim]
            mapped_labels_embd.append(class_embd)

        # Concatenate all averaged embeddings
        mapped_labels_embd = torch.cat(mapped_labels_embd)  # [batch_size, embedding_dim]

        # Return detached tensors (gradients won't flow through these in alignment loss)
        return visual_feats.detach(), mapped_labels_embd.detach()
```

## Trainer Class (from ChexnetTrainer.py)

```python
class ChexnetTrainer(object):
    """
    Main training class that orchestrates the entire training and evaluation process.
    Handles data loading, model training, validation, testing, and checkpoint management.
    """
    def __init__(self, args):
        """
        Initialize the trainer with model, optimizer, datasets, and training state.

        Args:
            args: Configuration object with all hyperparameters and paths
        """
        self.args = args

        # Set device (GPU if available, else CPU)
        self.device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

        # Load pre-computed text embeddings for disease classes
        self.textual_embeddings = np.load(args.textual_embeddings)

        # Initialize the model and move to device
        self.model = ZSLNetHybrid(self.args, self.textual_embeddings, self.device).to(self.device)

        # Initialize optimizer (Adam with default hyperparameters)
        self.optimizer = optim.Adam(
            self.model.parameters(),
            lr=self.args.lr,
            betas=(0.9, 0.999),
            eps=1e-08,
            weight_decay=1e-5
        )

        # Learning rate scheduler: reduces LR when validation loss plateaus
        self.scheduler = ReduceLROnPlateau(self.optimizer, factor=0.1, patience=5, mode='min')

        # Loss function (for BCE-only mode, though ZSL mode uses ranking loss)
        self.loss = torch.nn.BCELoss(size_average=True)

        # Training state tracking
        self.auroc_min_loss = 0.0  # AUROC achieved at minimum validation loss
        self.start_epoch = 1  # Starting epoch (1 for new training, >1 if resuming)
        self.lossMIN = float('inf')  # Minimum validation loss seen so far
        self.max_auroc_mean = float('-inf')  # Maximum mean AUROC seen so far
        self.best_epoch = 1  # Epoch with best AUROC

        # Track validation losses for plotting
        self.val_losses = []

        # Print model and training configuration
        # print(self.model)
        # print(self.optimizer)
        # print(self.scheduler)
        # print(self.loss)
        print(f'\n\nloaded imagenet weights {self.args.pretrained}\n\n\n')

        # Load checkpoints if specified (resume training or load weights)
        self.resume_from()  # Resume training from checkpoint (includes optimizer state)
        self.load_from()  # Load weights from checkpoint (weights only)

        # Initialize datasets and data loaders
        self.init_dataset()

        # Parse learning rate decay steps
        self.steps = [int(step) for step in self.args.steps.split(',')]

        # Time tracking for ETA calculation
        self.time_start = time.time()
        self.time_end = time.time()

        # Flag to determine if we should run test set evaluation
        self.should_test = False

        # Store class IDs that are loaded in the training dataset
        self.model.class_ids_loaded = self.train_dl.dataset.class_ids_loaded


    def __call__(self):
        self.train()

    def step_lr(self, epoch):
        step = self.steps[0]
        for index, s in enumerate(self.steps):
            if epoch < s:
                break
            else:
                step = s

        lr = self.args.lr * (0.1 ** (epoch // step))
        for param_group in self.optimizer.param_groups:
            param_group['lr'] = lr

    def load_from(self):
        if self.args.load_from is not None:
            checkpoint = torch.load(self.args.load_from)
            self.model.load_state_dict(checkpoint['state_dict'])
            print(f'loaded checkpoint from {self.args.load_from}')

    def resume_from(self):
        if self.args.resume_from is not None:
            checkpoint = torch.load(self.args.resume_from)
            self.model.load_state_dict(checkpoint['state_dict'])
            self.optimizer.load_state_dict(checkpoint['optimizer'])
            self.start_epoch = checkpoint['epoch'] + 1
            self.lossMIN = checkpoint['lossMIN']
            self.max_auroc_mean = checkpoint['max_auroc_mean']
            print(f'resuming training from epoch {self.start_epoch}')

    def save_checkpoint(self, prefix='best'):
        path = f'{self.args.save_dir}/{prefix}_checkpoint.pth.tar'
        torch.save(
            {
            'epoch': self.epoch,
            'state_dict': self.model.state_dict(),
            'max_auroc_mean': self.max_auroc_mean,
            'optimizer' : self.optimizer.state_dict(),
            'lossMIN' : self.lossMIN
            }, path)
        print(f"saving {prefix} checkpoint")
    def init_dataset(self):
        normalize = transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])

        train_transforms = []
        train_transforms.append(transforms.RandomResizedCrop(self.args.crop))
        train_transforms.append(transforms.RandomHorizontalFlip())
        train_transforms.append(transforms.ToTensor())
        train_transforms.append(normalize)

        datasetTrain = NIHChestXray(self.args, self.args.train_file, transform=transforms.Compose(train_transforms))

        self.train_dl = DataLoader(dataset=datasetTrain, batch_size=self.args.batch_size, shuffle=True,  num_workers=4, pin_memory=True)


        test_transforms = []
        test_transforms.append(transforms.Resize(self.args.resize))
        test_transforms.append(transforms.TenCrop(self.args.crop))
        test_transforms.append(transforms.Lambda(lambda crops: torch.stack([transforms.ToTensor()(crop) for crop in crops])))
        test_transforms.append(transforms.Lambda(lambda crops: torch.stack([normalize(crop) for crop in crops])))


        datasetVal =   NIHChestXray(self.args, self.args.val_file, transform=transforms.Compose(test_transforms))
        self.val_dl = DataLoader(dataset=datasetVal, batch_size=self.args.batch_size*10, shuffle=False, num_workers=4, pin_memory=True)

        datasetTest = NIHChestXray(self.args, self.args.test_file, transform=transforms.Compose(test_transforms), classes_to_load='all')
        self.test_dl = DataLoader(dataset=datasetTest, batch_size=self.args.batch_size*3, num_workers=8, shuffle=False, pin_memory=True)
        print(datasetTest.CLASSES)


    def train(self):
        """
        Main training loop.
        For each epoch:
        1. Train on training set
        2. Validate on validation set
        3. Save checkpoints if performance improves
        4. Test on test set if new best model found
        5. Update learning rate scheduler
        """
        for self.epoch in range(self.start_epoch, self.args.epochs):
            # Train for one epoch
            self.epochTrain()

            # Validate on validation set
            lossVal, val_ind_auroc = self.epochVal()
            val_ind_auroc = np.array(val_ind_auroc)

            # Compute mean AUROC across all classes
            aurocMean = val_ind_auroc.mean()

            # Always save the last epoch checkpoint
            self.save_checkpoint(prefix=f'last_epoch')
            self.should_test = False  # Reset test flag

            # Check if this is the best model by AUROC
            if aurocMean > self.max_auroc_mean:
                self.max_auroc_mean = aurocMean
                self.save_checkpoint(prefix='best_auroc')  # Save best AUROC model
                self.best_epoch = self.epoch
                self.should_test = True  # Test this model

            # Check if this is the model with minimum validation loss
            if lossVal < self.lossMIN:
                self.lossMIN = lossVal
                self.auroc_min_loss = aurocMean  # Track AUROC at min loss
                self.save_checkpoint(prefix='min_loss')  # Save min loss model
                self.should_test = True  # Test this model

            # Print validation AUROC results
            self.print_auroc(val_ind_auroc, self.val_dl.dataset.class_ids_loaded, prefix='val')

            # If we have a new best model, evaluate on test set
            if self.should_test is True:
                test_ind_auroc = self.test()
                test_ind_auroc = np.array(test_ind_auroc)

                # Write results to file
                self.write_results(val_ind_auroc, self.val_dl.dataset.class_ids_loaded,
                                 prefix=f'\n\nepoch {self.epoch}\nval', mode='a')
                # Write test results for seen classes (classes seen during training)
                self.write_results(test_ind_auroc[self.test_dl.dataset.seen_class_ids],
                                 self.test_dl.dataset.seen_class_ids, prefix='\ntest_seen', mode='a')
                # Write test results for unseen classes (zero-shot learning evaluation)
                self.write_results(test_ind_auroc[self.test_dl.dataset.unseen_class_ids],
                                 self.test_dl.dataset.unseen_class_ids, prefix='\ntest_unseen', mode='a')

                # Print test results
                self.print_auroc(test_ind_auroc[self.test_dl.dataset.seen_class_ids],
                               self.test_dl.dataset.seen_class_ids, prefix='\ntest_seen')
                self.print_auroc(test_ind_auroc[self.test_dl.dataset.unseen_class_ids],
                               self.test_dl.dataset.unseen_class_ids, prefix='\ntest_unseen')

            # Plot validation loss curve
            plot_array(self.val_losses, f'{self.args.save_dir}/val_loss')

            # Print training summary
            print(f'best epoch {self.best_epoch} best auroc {self.max_auroc_mean} loss {lossVal:.6f} auroc at min loss {self.auroc_min_loss:0.4f}')

            # Update learning rate based on validation loss
            self.scheduler.step(lossVal)


    def get_eta(self, epoch, iter):
        self.time_end = time.time()
        delta = self.time_end - self.time_start
        delta = delta * (len(self.train_dl) * (self.args.epochs - epoch) - iter)
        sec = timedelta(seconds=int(delta))
        d = (datetime(1,1,1) + sec)
        eta = f"{d.day-1} Days {d.hour}:{d.minute}:{d.second}"
        self.time_start = time.time()

        return eta



    def epochTrain(self):
        """
        Train the model for one epoch on the training set.
        Processes batches, computes loss, backpropagates, and updates weights.
        """
        self.model.train()  # Set model to training mode (enables dropout, batch norm updates, etc.)
        epoch_loss = 0

        # Iterate over training batches
        for batchID, (inputs, target) in enumerate(self.train_dl):
            # Move data to device (GPU or CPU)
            target = target.to(self.device)  # Labels: [batch_size, num_classes]
            inputs = inputs.to(self.device)  # Images: [batch_size, channels, height, width]

            # Forward pass: compute predictions and loss
            output, loss = self.model(inputs, target, self.epoch)

            # Backward pass: compute gradients
            self.optimizer.zero_grad()  # Clear gradients from previous iteration
            loss.backward()  # Compute gradients
            self.optimizer.step()  # Update model weights

            # Calculate estimated time to completion
            eta = self.get_eta(self.epoch, batchID)
            epoch_loss += loss.item()

            # Print progress every 10 batches
            if batchID % 10 == 9:
                print(f" epoch [{self.epoch:04d} / {self.args.epochs:04d}] eta: {eta:<20} "
                      f"[{batchID:04}/{len(self.train_dl)}] lr: \t{self.optimizer.param_groups[0]['lr']:0.4E} "
                      f"loss: \t{epoch_loss/batchID:0.5f}")



    def epochVal(self):
        """
        Validate the model on the validation set.
        Uses test-time augmentation (10 crops) and computes AUROC for each class.

        Returns:
            lossVal: Average validation loss
            aurocIndividual: List of AUROC scores for each class
        """
        self.model.eval()  # Set model to evaluation mode (disables dropout, batch norm uses running stats)

        lossVal = 0

        # Accumulate predictions and ground truth for all batches
        outGT = torch.FloatTensor().to(self.device)  # Ground truth labels
        outPRED = torch.FloatTensor().to(self.device)  # Model predictions

        # Iterate over validation batches (with progress bar)
        for i, (inputs, target) in enumerate(tqdm(self.val_dl)):
            with torch.no_grad():  # Disable gradient computation for efficiency
                target = target.to(self.device)
                inputs = inputs.to(self.device)

                # Handle test-time augmentation: inputs have shape [batch_size, n_crops, channels, height, width]
                # where n_crops=10 (TenCrop augmentation)
                varTarget = torch.autograd.Variable(target)  # Legacy Variable wrapper
                bs, n_crops, c, h, w = inputs.size()  # Extract dimensions

                # Reshape to process all crops: [batch_size*n_crops, channels, height, width]
                varInput = torch.autograd.Variable(inputs.view(-1, c, h, w).to(self.device))

                # Forward pass: model will average predictions across crops
                varOutput, losstensor = self.model(varInput, varTarget, n_crops=n_crops, bs=bs)

                # Accumulate predictions and ground truth
                outPRED = torch.cat((outPRED, varOutput), 0)  # [total_samples, num_classes]
                outGT = torch.cat((outGT, target), 0)  # [total_samples, num_classes]

                # Accumulate loss
                lossVal += losstensor.item()

                # Free memory
                del varOutput, varTarget, varInput, target, inputs

        # Compute average loss
        lossVal = lossVal / len(self.val_dl)

        # Compute AUROC for each class
        aurocIndividual = self.computeAUROC(outGT, outPRED, self.val_dl.dataset.class_ids_loaded)

        # Track validation loss for plotting
        self.val_losses.append(lossVal)

        return lossVal, aurocIndividual



    def test(self):
        """
        Test the model on the test set.
        Similar to validation but on the full test set (includes both seen and unseen classes).

        Returns:
            aurocIndividual: List of AUROC scores for each class in the test set
        """
        # Enable CUDA benchmarking for faster inference (optimizes convolution operations)
        cudnn.benchmark = True

        # Accumulate predictions and ground truth
        outGT = torch.FloatTensor().to(self.device)
        outPRED = torch.FloatTensor().to(self.device)

        self.model.eval()  # Set to evaluation mode

        # Iterate over test batches
        for i, (inputs, target) in enumerate(tqdm(self.test_dl)):
            with torch.no_grad():  # No gradients needed for testing
                target = target.to(self.device)
                outGT = torch.cat((outGT, target), 0)  # Accumulate ground truth

                # Handle test-time augmentation (10 crops)
                bs, n_crops, c, h, w = inputs.size()

                # Reshape to process all crops
                varInput = torch.autograd.Variable(inputs.view(-1, c, h, w).to(self.device))

                # Forward pass (no labels needed, just predictions)
                out, _ = self.model(varInput, n_crops=n_crops, bs=bs)

                # Accumulate predictions
                outPRED = torch.cat((outPRED, out.data), 0)

        # Compute AUROC for each class
        aurocIndividual = self.computeAUROC(outGT, outPRED, self.test_dl.dataset.class_ids_loaded)

        # Compute mean AUROC (for reference, not returned)
        aurocMean = np.array(aurocIndividual).mean()

        return aurocIndividual

    def computeAUROC(self, dataGT, dataPRED, class_ids):
        """
        Compute Area Under ROC Curve (AUROC) for each class.

        Args:
            dataGT: Ground truth labels [num_samples, num_classes]
            dataPRED: Predicted scores [num_samples, num_classes]
            class_ids: List of class IDs to compute AUROC for

        Returns:
            List of AUROC scores, one for each class in class_ids
        """
        outAUROC = []

        # Convert to numpy for sklearn
        datanpGT = dataGT.cpu().numpy()  # Ground truth: [num_samples, num_classes]
        datanpPRED = dataPRED.cpu().numpy()  # Predictions: [num_samples, num_classes]

        # Compute AUROC for each class
        for i in class_ids:
            # AUROC for binary classification: true labels vs predicted scores for class i
            auroc = roc_auc_score(datanpGT[:, i], datanpPRED[:, i])
            outAUROC.append(auroc)

        return outAUROC

    def write_results(self, aurocIndividual, class_ids, prefix='val', mode='a'):

        with open(f"{self.args.save_dir}/results.txt", mode) as results_file:

            aurocMean = aurocIndividual.mean()

            results_file.write(f'{prefix} AUROC mean {aurocMean:0.4f}\n')
            for i, class_id in enumerate(class_ids):
                results_file.write(f'{self.val_dl.dataset.CLASSES[class_id]} {aurocIndividual[i]:0.4f}\n')

    def print_auroc(self, aurocIndividual, class_ids, prefix='val'):
        aurocMean = aurocIndividual.mean()

        print (f'{prefix} AUROC mean {aurocMean:0.4f}')

        for i, class_id in enumerate(class_ids):
            print (f'{self.val_dl.dataset.CLASSES[class_id]} {aurocIndividual[i]:0.4f}')
```

```python
# Training function (from train.py)
def run_training():
    """
    Main training function.
    Sets random seeds, creates output directory, trains the model, and tests both saved checkpoints.

    Returns:
        trainer: The trained ChexnetTrainer object
    """
    # Set random seeds for reproducibility
    seed = 1002
    torch.manual_seed(seed)
    np.random.seed(seed)

    # Create output directory for checkpoints
    try:
        os.mkdir(args.save_dir)
    except OSError as error:
        print(error)  # Directory may already exist, that's okay

    # Initialize trainer and start training
    trainer = ChexnetTrainer(args)
    trainer()  # Calls trainer.train() which runs the full training loop

    # After training, test both saved checkpoints
    # Test 1: Model with minimum validation loss
    checkpoint = torch.load(f'{args.save_dir}/min_loss_checkpoint.pth.tar', weights_only=False)
    trainer.model.load_state_dict(checkpoint['state_dict'])
    print('Testing the min loss model')
    test_ind_auroc = trainer.test()
    test_ind_auroc = np.array(test_ind_auroc)

    # Print results for seen and unseen classes
    trainer.print_auroc(test_ind_auroc[trainer.test_dl.dataset.seen_class_ids],
                       trainer.test_dl.dataset.seen_class_ids, prefix='\ntest_seen')
    trainer.print_auroc(test_ind_auroc[trainer.test_dl.dataset.unseen_class_ids],
                       trainer.test_dl.dataset.unseen_class_ids, prefix='\ntest_unseen')

    # Test 2: Model with best AUROC
    checkpoint = torch.load(f'{args.save_dir}/best_auroc_checkpoint.pth.tar', weights_only=False)
    trainer.model.load_state_dict(checkpoint['state_dict'])
    print('Testing the best AUROC model')
    test_ind_auroc = trainer.test()
    test_ind_auroc = np.array(test_ind_auroc)

    # Print results for seen and unseen classes
    trainer.print_auroc(test_ind_auroc[trainer.test_dl.dataset.seen_class_ids],
                       trainer.test_dl.dataset.seen_class_ids, prefix='\ntest_seen')
    trainer.print_auroc(test_ind_auroc[trainer.test_dl.dataset.unseen_class_ids],
                       trainer.test_dl.dataset.unseen_class_ids, prefix='\ntest_unseen')

    return trainer

# Testing function (from test.py)
def run_test(checkpoint_path=None):
    """
    Test a trained model on the test set.
    Loads a checkpoint and evaluates on both seen and unseen classes.

    Args:
        checkpoint_path: Path to checkpoint file. If None, uses the best_auroc checkpoint.

    Returns:
        trainer: The ChexnetTrainer object with loaded model
        test_ind_auroc: Array of AUROC scores for each class
    """
    # Create output directory if it doesn't exist
    try:
        os.mkdir(args.save_dir)
    except OSError as error:
        print(error)  # Directory may already exist

    # Initialize trainer (this loads datasets)
    trainer = ChexnetTrainer(args)

    # Determine which checkpoint to load
    if checkpoint_path is None:
        checkpoint_path = f'{args.save_dir}/best_auroc_checkpoint.pth.tar'

    # Load checkpoint and update model weights
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    checkpoint = torch.load(checkpoint_path, map_location = device, weights_only=False)
    trainer.model.load_state_dict(checkpoint['state_dict'])
    print(f'Testing model from checkpoint: {checkpoint_path}')

    # Run test evaluation
    test_ind_auroc = trainer.test()
    test_ind_auroc = np.array(test_ind_auroc)

    # Print results for seen classes (classes available during training)
    trainer.print_auroc(test_ind_auroc[trainer.test_dl.dataset.seen_class_ids],
                       trainer.test_dl.dataset.seen_class_ids, prefix='\ntest_seen')
    # Print results for unseen classes (zero-shot learning evaluation)
    trainer.print_auroc(test_ind_auroc[trainer.test_dl.dataset.unseen_class_ids],
                       trainer.test_dl.dataset.unseen_class_ids, prefix='\ntest_unseen')

    return trainer, test_ind_auroc
```

# Early Stopping

```python
def run_training():
    """
    Main training function with early stopping.
    """
    # Set random seeds for reproducibility
    seed = 1002
    torch.manual_seed(seed)
    np.random.seed(seed)

    # Create output directory for checkpoints
    try:
        os.mkdir(args.save_dir)
    except OSError as error:
        print(error)

    # Initialize trainer with early stopping
    trainer = ChexnetTrainer(args)
    
    # Add early stopping parameters
    trainer.early_stop_patience = 5
    trainer.early_stop_counter = 0
    trainer.best_val_auroc = 0
    
    # Override train method with early stopping
    original_train = trainer.train
    def train_with_early_stopping():
        for trainer.epoch in range(trainer.start_epoch, trainer.args.epochs):
            trainer.epochTrain()
            lossVal, val_ind_auroc = trainer.epochVal()
            val_auroc_mean = np.array(val_ind_auroc).mean()
            # Test on new best models
            if val_auroc_mean > trainer.max_auroc_mean or lossVal < trainer.lossMIN:
                test_ind_auroc = trainer.test()
                test_ind_auroc = np.array(test_ind_auroc)
                trainer.print_auroc(test_ind_auroc[trainer.test_dl.dataset.seen_class_ids],
                                   trainer.test_dl.dataset.seen_class_ids, prefix='\ntest_seen')
                trainer.print_auroc(test_ind_auroc[trainer.test_dl.dataset.unseen_class_ids],
                                   trainer.test_dl.dataset.unseen_class_ids, prefix='\ntest_unseen')
            
            # Early stopping check
            if val_auroc_mean > trainer.best_val_auroc:
                trainer.best_val_auroc = val_auroc_mean
                trainer.early_stop_counter = 0
            else:
                trainer.early_stop_counter += 1
                
            if trainer.early_stop_counter >= trainer.early_stop_patience:
                print(f"Early stopping at epoch {trainer.epoch}")
                break
                
            # Continue with original training logic
            trainer.save_checkpoint(prefix='last_epoch')
            if val_auroc_mean > trainer.max_auroc_mean:
                trainer.max_auroc_mean = val_auroc_mean
                trainer.save_checkpoint(prefix='best_auroc')
            if lossVal < trainer.lossMIN:
                trainer.lossMIN = lossVal
                trainer.save_checkpoint(prefix='min_loss')
    
    trainer.train = train_with_early_stopping
    trainer()

    # Rest of function unchanged...
    checkpoint = torch.load(f'{args.save_dir}/min_loss_checkpoint.pth.tar', weights_only=False)
    trainer.model.load_state_dict(checkpoint['state_dict'])
    print('Testing the min loss model')
    test_ind_auroc = trainer.test()
    test_ind_auroc = np.array(test_ind_auroc)

    trainer.print_auroc(test_ind_auroc[trainer.test_dl.dataset.seen_class_ids],
                       trainer.test_dl.dataset.seen_class_ids, prefix='\ntest_seen')
    trainer.print_auroc(test_ind_auroc[trainer.test_dl.dataset.unseen_class_ids],
                       trainer.test_dl.dataset.unseen_class_ids, prefix='\ntest_unseen')

    checkpoint = torch.load(f'{args.save_dir}/best_auroc_checkpoint.pth.tar', weights_only=False)
    trainer.model.load_state_dict(checkpoint['state_dict'])
    print('Testing the best AUROC model')
    test_ind_auroc = trainer.test()
    test_ind_auroc = np.array(test_ind_auroc)

    trainer.print_auroc(test_ind_auroc[trainer.test_dl.dataset.seen_class_ids],
                       trainer.test_dl.dataset.seen_class_ids, prefix='\ntest_seen')
    trainer.print_auroc(test_ind_auroc[trainer.test_dl.dataset.unseen_class_ids],
                       trainer.test_dl.dataset.unseen_class_ids, prefix='\ntest_unseen')

    return trainer
```
